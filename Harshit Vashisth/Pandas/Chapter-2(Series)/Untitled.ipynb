```python
# 2010-2014
# males,feamales,switzerland
mask1=df.year.between(2010,2014)
mask2=df.sex=='male'
mask3=df.country=='Switzerland'
mask4=df.age=='15-24 years'
f_df=df.loc[mask1 & mask2 & mask3 & mask4,['country','age','sex','suicides_no']]
```

```python
f_df
```

```python
f_df.sum().suicides_no
```
